### ERROR BAR

```python
mean_cases = data['cases'].mean()
mean_deaths = data['deaths'].mean()
std_cases = data['cases'].std()
std_deaths = data['deaths'].std()

# Create an error plot with mean and standard deviation as data points
plt.figure(figsize=(8, 6))
plt.errorbar(['Mean Cases', 'Mean Deaths'], [mean_cases, mean_deaths], yerr=[std_cases, std_deaths], fmt='o', color='blue')
plt.xlabel('COVID-19 DEATH')
plt.ylabel('COVID-19 CASES')
plt.title('ERROR PLOTS')
plt.grid()

# Print the calculated mean and standard deviation
print("Mean Cases:", mean_cases)
print("Mean Deaths:", mean_deaths)
print("Standard Deviation of Cases:", std_cases)
print("Standard Deviation of Deaths:", std_deaths)

plt.show()
```

### HISTOGRAM

```python
data = np.random.randn(2000)  

# Create a histogram
plt.figure(figsize=(8, 6))
plt.hist(data, bins=20, color='blue', alpha=0.7, edgecolor='black')
plt.xlabel('COVID-19 CASES')
plt.ylabel('COVID-19 DEATH')
plt.title('HISTOGRAM')

# Calculate mean and standard deviation
mean_value = np.mean(data)
std_deviation = np.std(data)

# Add mean and standard deviation to the plot
plt.axvline(mean_value, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean_value:.2f}')
plt.axvline(mean_value + std_deviation, color='green', linestyle='dashed', linewidth=2, label=f'Std Dev: {std_deviation:.2f}')
plt.axvline(mean_value - std_deviation, color='green', linestyle='dashed', linewidth=2)

# Add legend
plt.legend()

# Display the plot
plt.show()

print(f"Mean: {mean_value:.2f}")
print(f"Standard Deviation: {std_deviation:.2f}")
```

### INFERENCE
We have generated the  dataset with 1000 data points. Further We created a histogram of the data using plt.hist(), specifying the number of bins. The density=True parameter normalizes the histogram and calculated the mean and standard deviation of the dataset using NumPy's np.mean() and np.std() functions.
